```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
plt.title('Price is US$')

list_country = ['Italy', 'Mexico', 'Canada', 'Japan', 'USA', 'Germany']

for country in gas:
    if country in list_country:
        plt.plot(gas['Year'], gas[country], label = [country], marker = '.')

plt.xlabel('Years')
plt.ylabel('US$')

plt.legend()

plt.show()
```
